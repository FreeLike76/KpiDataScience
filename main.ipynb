## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Configs

```python
np.random.seed(6)
plt.rcParams["figure.figsize"] = (20, 6)

SELECTION_SIZE = 1000
```

## Errors

### Normal Error

```python
normal_mean = 2
normal_median = normal_mean
normal_std = 2

normal_error = np.random.normal(normal_mean, normal_std, SELECTION_SIZE)

print("Expected mean:", normal_mean)
print("Expected median:", normal_median)
print("Expected std:", normal_std)
print("\n")
print("Real mean:", np.mean(normal_error))
print("Real median:", np.median(normal_error))
print("Real std:", np.std(normal_error))
print("\n")
print(normal_error[:25])
```

### Normal Error Plot

```python
plt.hist(normal_error, bins=20, alpha=0.5, label="normal_error")
plt.show()
```

### Exponential Error

```python
exp_lambda = 0.5
exp_beta = 1 / exp_lambda

exp_mean = 1 / exp_lambda
exp_median = math.log(2) / exp_lambda
exp_std = math.sqrt(1 / (exp_lambda ** 2))

exp_error = np.random.exponential(exp_beta, SELECTION_SIZE)

print("Expected mean:", exp_mean)
print("Expected median:", exp_median)
print("Expected std:", exp_std)
print("\n")
print("Real mean:", np.mean(exp_error))
print("Real median:", np.median(exp_error))
print("Real std:", np.std(exp_error))
print("\n")
print(exp_error[:25])
```

### Exponential Error Plot

```python
plt.hist(exp_error, bins=20, alpha=0.5, label="normal_error")
plt.show()
```

## Creating models

### Linear Model

```python
linear_model = np.linspace(0, 5, SELECTION_SIZE, dtype=np.float32)
print(linear_model[:25])
```

### Linear Model Plot

```python
plt.plot(np.arange(SELECTION_SIZE), linear_model)
plt.show()
```

### Quadratic model

```python
quadratic_model = np.square(np.linspace(0, 5, SELECTION_SIZE, dtype=np.float32))
print(quadratic_model[:25])
```

### Quadratic Model Plot

```python
plt.plot(np.arange(SELECTION_SIZE), quadratic_model)
plt.show()
```

## Combining model and error

### Linear With Errors

```python
linear_model_normal = linear_model + normal_error
linear_model_exp = linear_model + exp_error

plt.plot(np.arange(SELECTION_SIZE), linear_model, label="linear")
plt.plot(np.arange(SELECTION_SIZE), linear_model_normal, label="linear + normal")
plt.plot(np.arange(SELECTION_SIZE), linear_model_exp, label="linear + exp")
plt.legend()
plt.show()
```

```python
plt.hist(normal_error, bins=25, alpha=0.5, label="normal")
plt.hist(linear_model_normal, bins=25, alpha=0.5, label="linear + normal")
plt.legend()
plt.show()
```

```python
plt.hist(exp_error, bins=50, alpha=0.5, label="exp")
plt.hist(linear_model_exp, bins=50, alpha=0.5, label="linear + exp")
plt.legend()
plt.show()
```

### Quadratic With Errors

```python
quadratic_model_normal = quadratic_model + normal_error
quadratic_model_exp = quadratic_model + exp_error

plt.plot(np.arange(SELECTION_SIZE), quadratic_model, label="quadratic")
plt.plot(np.arange(SELECTION_SIZE), quadratic_model_normal, label="quadratic + normal")
plt.plot(np.arange(SELECTION_SIZE), quadratic_model_exp, label="quadratic + exp")
plt.legend()
plt.show()
```

```python
plt.hist(normal_error, bins=50, alpha=0.5, label="normal")
plt.hist(quadratic_model_normal, bins=50, alpha=0.5, label="quadratic + normal")
plt.legend()
plt.show()
```

```python
plt.hist(exp_error, bins=50, alpha=0.5, label="exp")
plt.hist(quadratic_model_exp, bins=50, alpha=0.5, label="quadratic + exp")
plt.legend()
plt.show()
```
