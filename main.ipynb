## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Configs

```python
np.random.seed(6)
plt.rcParams["figure.figsize"] = (20, 6)
```

## Creating models

### Model params

```python
SELECTION_SIZE = 1000
```

### Linear Model

```python
linear_model = np.linspace(0, 100, SELECTION_SIZE, dtype=np.float32)
linear_model
```

### Linear Model Plot

```python
plt.plot(np.arange(SELECTION_SIZE), linear_model)
plt.show()
```

### Quadratic model

```python
quadratic_model = np.square(np.linspace(0, 10, SELECTION_SIZE, dtype=np.float32))
print(quadratic_model)
```

### Quadratic Model Plot

```python
plt.plot(np.arange(SELECTION_SIZE), quadratic_model)
plt.show()
```

## Errors

### Normal Error

```python
normal_mean = 0
normal_std = 5
normal_error = np.random.normal(normal_mean, normal_std, SELECTION_SIZE)
print(normal_error)
```

### Normal Error Plot

```python
plt.hist(normal_error, bins=20, alpha=0.5, label="normal_error")
plt.show()
```

### Exponential Error

```python
exp_error = np.exp(np.linspace(0, 10, SELECTION_SIZE)).shu
print(exp_error)
```
